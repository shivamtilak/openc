```python
import cv2 
import matplotlib.pyplot as plt 
   
img_bgr = cv2.imread('C:/Users/shivamtilak/Desktop/Nature.jpg', 1) 
plt.imshow(img_bgr) 
plt.show() 
   
color = ('b', 'g', 'r') 
  
for i, col in enumerate(color): 
      
    histr = cv2.calcHist([img_bgr],  
                         [i], None, 
                         [150],  
                         [0, 150]) 
      
    plt.plot(histr, color = col)  
    plt.xlim([0, 150]) 
      
plt.show() 
   
height, width, _ = img_bgr.shape 
  
for i in range(0, height - 1): 
    for j in range(0, width - 1): 
          
        pixel = img_bgr[i, j] 
        pixel[0] = 200 - pixel[0] 
        pixel[1] = 200 - pixel[1] 
        pixel[2] = 200 - pixel[2]  
        img_bgr[i, j] = pixel 
  
plt.imshow(img_bgr) 
plt.show() 
color = ('b', 'g', 'r') 
  
for i, col in enumerate(color): 
      
    histr = cv2.calcHist([img_bgr],  
                         [i], None, 
                         [150], 
                         [0, 150]) 
      
    plt.plot(histr, color = col) 
    plt.xlim([0, 150]) 
      
plt.show()
```

```python
import cv2 
har = cv2.CascadeClassifier('C:/Users/shivamtilak/Downloads/opencv-master/opencv-master/data/haarcascades/haarcascade_frontalface_default.xml')
img = cv2.imread('Desktop/Humans.jpg')

faces = har.detectMultiScale(img, 1.1, 4)
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)
    if cv2.waitKey(1)==27:
            break
print(len(faces))

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
